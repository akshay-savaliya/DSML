```python
import pandas as pd
```

```python
people = {
    "first": ["Kirtan", "Anirudh", "Darshit"],
    "last": ["patel", "rana", "patel"],
    "email": ["kirtanpatel@gmail.com", "Anirudhr@gmail.com", "darshitpatel@gmail.com"]
}
```

```python
df = pd.DataFrame(people)
df
```

```python
df['last'] == 'patel'
```

```python
# What if we want to get the dataframe for last == patel
```

```python
df.loc[df['last'] == 'patel']
```

```python
filt = (df['last'] == 'patel')
```

```python
# the same can be accessed using loc
df.loc[filt]
```

```python
# If we need only email from filtered rows then
df.loc[filt, 'email']
```

```python
df.loc[filt, ['first', 'email']]
```

### AND and OR in Pandas Filtering
#### AND- &
#### OR - |

```python
# And
filt = ((df['last'] == 'patel') & (df['first'] == 'Darshit'))
```

```python
df.loc[filt]
```

```python
# OR
filt = ((df['last'] == 'rana') | (df['first'] == 'Darshit'))
```

```python
df.loc[filt]
```

#### Not - ~

```python
df.loc[~filt]
```
