```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.style.use("fivethirtyeight")
```

### Histograms are used to represent the continuous distribution of values over multiple bins

```python
ages = [18, 19, 21, 25, 26, 26, 30, 32, 38, 45, 55]

plt.hist(ages)

plt.title('Ages of Students')
plt.xlabel('Ages')
plt.ylabel('Total Students')
```

#### Lets provide bins explicitly 

```python
ages = [18, 19, 21, 25, 26, 26, 30, 32, 38, 45, 55]

plt.hist(ages, bins = 5)


plt.title('Ages of Students')
plt.xlabel('Ages')
plt.ylabel('Total Students')
```

```python
ages = [18, 19, 21, 25, 26, 26, 30, 32, 38, 45, 55]

plt.hist(ages, bins = 5, edgecolor = 'black')


plt.title('Ages of Students')
plt.xlabel('Ages')
plt.ylabel('Total Students')
```

#### Lets give our list of bins

```python
ages = [18, 19, 21, 25, 26, 26, 30, 32, 38, 45, 55]
bins = [10, 20, 30, 40, 50 ,60]
plt.hist(ages, bins = bins, edgecolor = 'black')


plt.title('Ages of Students')
plt.xlabel('Ages')
plt.ylabel('Total Students')
```

#### What if we provide the bins which are not covering all values

```python
ages = [18, 19, 21, 25, 26, 26, 30, 32, 38, 45, 55]
bins = [20, 30, 40, 50 ,60]
plt.hist(ages, bins = bins, edgecolor = 'black')


plt.title('Ages of Students')
plt.xlabel('Ages')
plt.ylabel('Total Students')
```

#### Vertical Line for Median Age

```python
ages = [18, 19, 21, 25, 26, 26, 30, 32, 38, 45, 55]
bins = [10, 20, 30, 40, 50 ,60]
median_age = 27

plt.hist(ages, bins = bins, edgecolor = 'black')

plt.axvline(median_age, color='red', label='Median Age')

plt.title('Ages of Students')
plt.xlabel('Ages')
plt.ylabel('Total Students')
plt.legend()
```

#### Line width can be customized

```python
ages = [18, 19, 21, 25, 26, 26, 30, 32, 38, 45, 55]
bins = [10, 20, 30, 40, 50 ,60]
median_age = 27

plt.hist(ages, bins = bins, edgecolor = 'black')

plt.axvline(median_age, color='red', label='Median Age', linewidth= 2)

plt.title('Ages of Students')
plt.xlabel('Ages')
plt.ylabel('Total Students')
```
