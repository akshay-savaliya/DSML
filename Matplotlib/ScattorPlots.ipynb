### Scatter Plots
#### Used when relationship between 2 sets of values to be compared and to see if they are corelated

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.style.use('seaborn')
```

```python
x = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
y = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 5, 1]
```

```python
plt.scatter(x,y)
```

#### Changing the size of dots

```python
plt.scatter(x,y, s=100)
```

#### Changing color

```python
plt.scatter(x,y, s=100, c='green')
```

#### Instead of dots we can use other markers like X or o

```python
plt.scatter(x,y, s=100, c='green', marker ='x')
```

#### Giving Edgecolor

```python
plt.scatter(x,y, c='green', edgecolor='black', linewidth=1, alpha = 0.75)
```

#### Adding Additional Information to dataset using color

```python
colors = [7, 5, 9, 7, 5, 7, 2, 5, 3, 7, 1, 2, 8, 1, 9, 2, 5, 6, 7, 5]
plt.scatter(x,y, s=100, c=colors, edgecolor='black', linewidth=1, alpha=0.75)
```

#### Shades of grey can be changed here 

```python
x = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
y = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 5, 1]

colors = [7, 5, 9, 7, 5, 7, 2, 5, 3, 7, 1, 2, 8, 1, 9, 2, 5, 6, 7, 5]
plt.scatter(x,y, s=100, c=colors, cmap = 'Greens', edgecolor='black', linewidth=1)

cbar = plt.colorbar()
cbar.set_label('Rating')
```

### Lets work on real data

```python
import pandas as pd
```

```python
data = pd.read_csv('Scatter Plot Data.csv')
```

```python
view_count = data['view_count']
likes = data['likes']
ratio = data['ratio']
```

```python
plt.scatter(view_count, likes, s=100, edgecolor='black', linewidth=1, alpha = 0.75)
```

#### Not Looking a good plot because of an outlier

#### Hence lets scale the plot to log scale

```python
plt.scatter(view_count, likes, edgecolor='black', linewidth=1, alpha=0.75)

plt.xscale('log')
plt.yscale('log')
```

#### Lets use ratio of likes/dislikes to represent more information in plot using colors

```python
plt.scatter(view_count, likes, c=ratio, cmap = 'summer', edgecolor='black', linewidth=1, alpha=0.75)

cbar = plt.colorbar()
cbar.set_label('Like/Dislike Ratio')

plt.xscale('log')
plt.yscale('log')
```
