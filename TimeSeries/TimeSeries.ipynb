```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import pandas as pd
from datetime import datetime, timedelta
import matplotlib.dates as mpl_dates
```

```python
plt.style.use('seaborn')
```

#### First Time Series Plot

```python
dates = [
    datetime(2021, 6, 10),
    datetime(2021, 6, 11),
    datetime(2021, 6, 12),
    datetime(2021, 6, 13),
    datetime(2021, 6, 14),
    datetime(2021, 6, 15),
    datetime(2021, 6, 16)
]

y = [0, 1, 3, 4, 6, 5, 7]


plt.plot_date(dates, y)
```

#### Lets connect the dots with line

```python
plt.plot_date(dates, y, linestyle = 'solid')
```

#### Giving titles and labels

```python
plt.plot_date(dates, y, linestyle = 'solid')

plt.title('Share Price of XYZ')
plt.xlabel('Date')
plt.ylabel('Closing Price')
```

#### Foramtting Date & Time

```python
plt.plot_date(dates, y, linestyle = 'solid')

date_format = mpl_dates.DateFormatter('%b, %d %Y')
plt.gca().xaxis.set_major_formatter(date_format)

plt.title('Share Price of XYZ')
plt.xlabel('Date')
plt.ylabel('Closing Price')
```

#### Dates are not looking properly 

```python
plt.plot_date(price_date, price_close, linestyle = 'solid')

plt.gcf().autofmt_xdate()

plt.title('Share Price of XYZ')
plt.xlabel('Date')
plt.ylabel('Closing Price')
```

#### So here the column price_date is not the actual date, but String
#### so lets convert it to date using pandas method

```python
data = pd.read_csv('TimeSeries Data.csv')

data['Date'] = pd.to_datetime(data['Date'])
data.sort_values('Date', inplace= True)

price_date = data['Date']
price_close = data['Close']

plt.plot_date(price_date, price_close, linestyle = 'solid')

plt.gcf().autofmt_xdate()

plt.title('Share Price of XYZ')
plt.xlabel('Date')
plt.ylabel('Closing Price')
```
